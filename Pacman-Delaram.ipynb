```python
#Creating a dictionary to map states into indexes and vice versa

state_to_index = {}
index_to_state = {}

cnt = 0

for C1 in ['E', 'D', 'W']:
    for C2 in ['E', 'D', 'W']:
        for C3 in ['E', 'D', 'W']:
            for C4 in ['E', 'D', 'W']:
                C5 = 'A'
                for C6 in ['E', 'D', 'W']:
                    for C7 in ['E', 'D', 'W']:
                        for C8 in ['E', 'D', 'W']:
                            for C9 in ['E', 'D', 'W']:
                                state = C1 + C2 + C3 + C4 + C5 + C6 + C7 + C8 + C9
                                print(state)
                                state_to_index[state] = cnt
                                index_to_state[cnt] = state
                                cnt = cnt + 1
```

```python
#Creating delta function
def delta(state, action, xA, yA, mapp):
    reward = 0
#     Calculate where we will be with this action
    next_yA = yA + (action == 'Right') - (action == 'Left')
    next_xA = xA + (action == 'Down') - (action == 'Up')
    
#     If we encounter a wall, reward -1, and also dont change the position
    if(mapp[next_xA][next_yA] == 'W'):
        reward -= 1
    else:
#         if encountered dot, reward +1
        if(mapp[next_xA][next_yA] == 'D'):
            reward += 1
#         Change the previous agent's position to empty
#         Change agents position as the action said 
        mapp[xA] = mapp[xA][:yA] + 'E' + mapp[xA][yA + 1:]
        xA = next_xA
        yA = next_yA
        mapp[xA] = mapp[xA][:yA] + 'A' + mapp[xA][yA + 1:]
        
#   Update the state using our map  
    state = mapp[xA - 1][yA - 1:yA + 2] + mapp[xA][yA - 1:yA + 2] + mapp[xA + 1][yA - 1:yA + 2]
    
    return reward, state, mapp, xA, yA
```

```python
# Setting hypervalues and other variables
import numpy as np

action_types = ['Up', 'Down', 'Left', 'Right']
Q = np.zeros((len(state_to_index.keys()), len(action_types)))
e = 0.8
alpha = 0.1
gamma = 0.9
max_episode = 5000
max_episode_length = 80

```

```python
#Learning Q function

# Run Qlearning function with different gammas. Dont forget to redefine Q table.
for gamma in [0.25, 0.5, 1]:
    e = 0.8 # The probability of random choosing vs choosing the best choice till now for next state
    Q = np.zeros((len(state_to_index.keys()), len(action_types)))
    average_episode_reward = 0
    
#     Repeat learning for many episodes

    for i in range(max_episode):
        
#         initializing for first state

        mapp = ['WWWWWWWWWWW',
                'WADDDDDDDDW',
                'WDWWWDWWWDW',
                'WDWDDDDDWDW',
                'WDDDWEWDDDW',
                'WDWDWEWDWDW',
                'WDWDDWDDWDW',
                'WDDDDDDDDDW',
                'WWWWWWWWWWW']
        xA = 1
        yA = 1
        state = mapp[xA - 1][yA - 1:yA + 2] + mapp[xA][yA - 1:yA + 2] + mapp[xA + 1][yA - 1:yA + 2]


# Continue until you meet max length (avoiding walking in empty block too much)

        for j in range(max_episode_length):
#         Choose randomly with probabilty of e
            if np.random.rand() < e:
                action_index = np.random.randint(len(action_types))
            else:
                action_index = np.argmax(Q[state_to_index[state], :])

#           Calculate delta
            reward, next_state, mapp, xA, yA = delta(state, action_types[action_index], xA, yA, mapp)
            
#           Update Q
            Q[state_to_index[state], action_index] = (1 - alpha) * Q[state_to_index[state], action_index] + alpha * (reward + gamma * np.max(Q[state_to_index[next_state], :]))

            average_episode_reward = average_episode_reward + reward

            state = next_state
            
#           Check for meeting the goal
            have_D = sum([('D' in word) for word in mapp])

            if not have_D:
                break;

        if(i % 100 == 0):
            if(i < 800):
#        Every now and then reduce e, to give the values of Q more chance because they're now more stable and trustworthy
                e -= 0.1
        if (i % 1000 == 0):
            print(average_episode_reward / 1000)
            average_episode_reward = 0
    print("------------------------------------------------")
```
